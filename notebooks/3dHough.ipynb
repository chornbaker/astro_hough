```python
class NEO:
    def __init__(self,  xmin, xmax, ymin, ymax, maxR, vmin, vmax, T0):
        self.data = {}
        self.variance( xmin, xmax, ymin, ymax)
        self.direction()
        self.speed(vmin, vmax)
        self.radius(maxR)
        self.position( xmin, xmax, ymin, ymax)
        self.start_time = T0
    def position(self,  xmin, xmax, ymin, ymax):
        self.data['x_curr'] = np.random.uniform(low=xmin, high=xmax)
        self.data['y_curr'] = np.random.uniform(low=ymin, high=ymax)
        self.data['x'] = [self.data['x_curr']]
        self.data['y'] = [self.data['y_curr']]
    def direction(self):
        self.theta = np.random.uniform(low=-np.pi/3., high=np.pi/3.)

    def speed(self, vmin, vmax):
        self.velocity = np.random.uniform(low=vmin, high=vmax)
    def variance(self,  xmin, xmax, ymin, ymax):
        self.var = np.random.uniform(low=0, high=(((xmin-xmax)**2+(ymin-ymax)**2)**0.5))
    def update_pos(self, Tcurrent):
        tstep = Tcurrent-self.start_time
        self.data['x_curr'] = self.data['x'][0]+self.velocity*np.cos(self.theta)*tstep
        self.data['y_curr'] = self.data['y'][0]+self.velocity*np.sin(self.theta)*tstep
        self.update_track()
    def update_track(self):
        self.data['x'].append(self.data['x_curr'])
        self.data['y'].append(self.data['y_curr'])
    def radius(self, maxR):
        self.data['radius_int'] = np.random.uniform(low=0, high=maxR)
        self.update_radius()      
    def update_radius(self):
        self.data['radius_curr'] = self.data['radius_int']*np.random.random()
    def get_track(self):
        return self.data['x'], self.data['y']
    def get_pos(self, smooth=100000000.):
        return self.data['x_curr']+ np.random.normal(loc=0, scale=self.var/smooth), self.data['y_curr'] + np.random.normal(loc=0, scale=self.var/smooth)
    def get_radius(self, tol=1e-5):
        if np.random.rand()<tol:
            return 0.
        else:
            return self.data['radius_curr']
```

```python
class reduction_object:
    def __init__(self, x, y, time,  xmin,xmax, ymin, ymax, epoch):
        self.data = {}
        self.data['epoch'] = epoch
        self.data['x'] = x
        self.data['y'] = y
        self.data['t'] = time
        self.data['P1'] = ((x-xmin)**2.+(y-ymin)**2.)**0.5
        self.data['P2'] = ((x-xmax)**2.+(y-ymax)**2.)**0.5
        self.status = True
    def get_P1(self):
        return self.data['P1']
    def get_P2(self):
        return self.data['P2']
    def get_epoch(self):
        return self.data['epoch']
    def get_x(self):
        return self.data['x']
    def get_y(self):
        return self.data['y']
    def update_status(self, boolean):
        self.status = boolean
    def get_time(self):
        return self.data['t']
    def get_status(self):
        return self.status
class generate_RO:
    def __init__(self, X, Y, time, xmin, xmax, ymin, ymax, epoch):
        self.ROs = []
        for i in xrange(len(X)):
            self.ROs.append(reduction_object(X[i], Y[i], time, xmin, xmax, ymin, ymax, epoch))
    def get_ROs(self):
        return self.ROs
```

##Basic Hough:

```python
def reduction(all_data, JulianDates,  num_epoch, xmin, xmax, ymin, ymax, tol1=0.00025,second_binnig=100,num_bins=500):
    sums = np.zeros((num_bins,num_bins))
    processed_data = np.zeros((num_bins,num_bins, num_epoch))
    for i in xrange(num_epoch):
        processed_data[:, :, i] = np.array(all_data[i])
        sums += np.array(all_data[i])
    sums = sums/float(num_epoch)
    
    filtered_data = np.zeros((num_bins,num_bins, num_epoch))
    for i in xrange(num_epoch):
        datum = processed_data[:,:,i]
        flatten = datum.reshape(num_bins**2, ) - sums.reshape(num_bins**2, )
        reduction = np.clip(flatten, a_min=0, a_max=np.inf)
        filtered_data[:,:,i] = reduction.reshape((num_bins,num_bins)).T
    
    reduced_data = []
    delta_x = (xmax-xmin)/float(num_bins-1)
    delta_y = (ymax-ymin)/float(num_bins-1)
    for i in xrange(num_epoch):
        reduced_xs = []
        reduced_ys = []
        datum =  filtered_data[:,:,i]
        N, M = datum.shape
        for n in xrange(N):
            for m in xrange(M):
                if datum[n,m]> 0:
                    reduced_xs.append(n*delta_x+xmin)
                    reduced_ys.append(m*delta_y+ymin)
        reduced_data.append((reduced_ys, reduced_xs))
    reduced_data = np.array(reduced_data) 
    
    All_ROs = []
    for epoch in xrange(num_epoch):
        Xepoch, Yepoch = reduced_data[epoch]
        RO = generate_RO(Xepoch, Yepoch, JulianDates[epoch], xmin, xmax, ymin, ymax, epoch).get_ROs()
        All_ROs.append(RO)
        
    for RO_set in All_ROs:
        points = []
        for RO in RO_set:
            points.append((RO.get_P1(), RO.get_P2()))
        sorted_points = sorted(points,key=itemgetter(0))
        x, y = zip(*sorted_points)    
        
    hist_info = np.histogram(x, bins=second_binnig)
    
    dict_set = []
    for RO_set in All_ROs:
        RO_dict = {}
        for i in xrange(second_binnig):
            RO_dict[str(i)] = []
        for RO in RO_set:
            distance = RO.get_P1()
            test = np.histogram(distance, bins=hist_info[1])
            RO_dict[str(np.argmax(test[0]))].append(RO)
        dict_set.append(RO_dict)
        
    keys = dict_set[0].keys()
    for key in keys:
        All_ROs = []
    
        for dictionary in dict_set:
            ROs = All_ROs[:]
            All_ROs = ROs+dictionary[key]
        Remaining_ROs = All_ROs[:]
        temp_dict = {}
        for epoch in xrange(num_epoch):
            temp_dict[str(epoch)] = []
        #split by epoch:
        for RO in Remaining_ROs:
            i = RO.get_epoch()
            temp_dict[str(i)].append(RO)
       
        for epoch1 in xrange(num_epoch):
            current_ROs = temp_dict[str(epoch1)]
            for RO1 in current_ROs:
                x1 = RO1.get_x()
                y1 = RO1.get_y()
                i1 = RO1.get_epoch()
                for epoch2 in xrange(num_epoch):
                    if epoch1 != epoch2:
                        comparing_ROs = temp_dict[str(epoch2)]
                        for RO2 in comparing_ROs:
                            x2 = RO2.get_x()
                            y2 = RO2.get_y()
                            i2 = RO2.get_epoch()
                            if(((x1-x2)**2+(y1-y2)**2)**0.5 < tol1): 
                                RO2.update_status(False)
                                RO1.update_status(False)
     
                            
    possibles = []
    for key in keys:
        for dictionary in dict_set:
            ROs = dictionary[key]
            for RO in ROs:
                if RO.get_status():
                    possibles.append(RO)
    return possibles
```

```python
%%time
def generate_HoughSpace(data, num_epoch,num_degrees =720):
    epoch_dict ={}
    thetas = np.linspace(-np.pi/2, np.pi/2, num_degrees)
    all_rhos = []
    all_thetas = []
    for i in xrange(num_epoch):
        epoch_dict[str(i)] = []
    for RO in data:
        i = RO.get_epoch()
        epoch_dict[str(i)].append(RO)
    
    for i in xrange(num_epoch):
        current_epoch = epoch_dict[str(i)]
        current_rhos = []
        current_thetas = []
        for RO in current_epoch:
            x =RO.get_x()
            y =RO.get_y()
            for theta in thetas:
                current_rhos.append(x*np.cos(theta)+y*np.sin(theta))
                current_thetas.append(theta)
        all_rhos.append(current_rhos)
        all_thetas.append(current_thetas)
    return np.array(all_rhos), np.array(all_thetas)
```

```python
def generate_possible_lines(data ,x_Bin_edges , y_Bin_edges,number_points):
    possible_line = []
    space = data.copy()
    N, M = space.shape
    space = space.reshape((N*M,))
    for k in xrange(number_points):
        max_i = np.argmax(space)
        i = max_i/N
        j = max_i - M*i
        space[max_i] = 0
        rho_1 = x_Bin_edges[i+1]
        theta_1 = y_Bin_edges[j+1]
        rho_2 = x_Bin_edges[i-1]
        theta_2 = y_Bin_edges[j-1]
        possible_line.append((rho_1, theta_1, rho_2, theta_2))
    return possible_line
```

```python
def generate_line(rho, theta, x_Vec):
    y_Vec = (rho - x_Vec*np.cos(theta))/np.sin(theta)
    return y_Vec
```

```python
def randomized_HS(ROs, num_epoch, num_pairs, tol = 0):
    #generate mapping epoch:ro
    epoch_dict ={}
    all_rhos = []
    all_thetas = []
    for i in xrange(num_epoch):
        epoch_dict[str(i)] = []
 
    for RO in ROs:
        i = RO.get_epoch()
        epoch_dict[str(i)].append(RO)
    rhos = []
    thetas = []
    count = 0

    while count < num_pairs:
        epoch1 = np.random.randint(low=0, high=num_epoch-1)
        epoch2 = np.random.randint(low=0, high=num_epoch-1)
        while epoch2 == epoch1:
            epoch2 = np.random.randint(low=0, high=num_epoch-1)
        epoch_ro1 = epoch_dict[str(epoch1)]
        epoch_ro2 = epoch_dict[str(epoch2)]
        idx1 = np.random.randint(low=0, high=len(epoch_ro1))
        idx2 = np.random.randint(low=0, high=len(epoch_ro2))
        RO1 = epoch_ro1[idx1]
        RO2 = epoch_ro2[idx2]
        x1 = RO1.get_x()
        y1 = RO1.get_y()
        x2 = RO2.get_x()
        y2 = RO2.get_y()
        D = ((x1-x2)**2 + (y1-y2)**2)**0.5
        
        if D > tol:
            if x2 == x1:
                theta = np.pi/2.
            if x2 != x1:
                m = (y2-y1)/(x1-x2)
                theta = np.arctan(m)#+np.pi/2.
            rho = x1*np.sin(theta) + y1*np.cos(theta)
            rhos.append(rho)
            thetas.append(theta)
            count += 1
    return rhos, thetas
```

```python
def PseudoLeSq_HS(ROs, num_epoch):
    epoch_dict ={}
    space = []
    for i in xrange(num_epoch):
        epoch_dict[str(i)] = []
    for RO in ROs:
        i = RO.get_epoch()
        epoch_dict[str(i)].append(RO)
    for i in xrange(num_epoch-1):
        current_epoch = epoch_dict[str(i)]
        
        for j in xrange(num_epoch):
            if j > i:
                compare_epoch =  epoch_dict[str(j)]
                for RO1 in current_epoch:
                    x1 = RO1.get_x()
                    y1 = RO1.get_y()
                    ep1 = RO1.get_epoch()
                    for RO2 in compare_epoch:
                        
                        diffs = []
                        x2 = RO2.get_x()
                        y2 = RO2.get_y()
                        ep2 = RO2.get_epoch()
                        m = np.divide((y2-y1), (x2-x1))
                        b = y1 - x1*m
                        D = ((y1-y2)**2.+(x1-x2)**2.)**0.5
                        t = ep2-ep1
                        for k in xrange(num_epoch):
                            if k > j:
                                final_epoch = epoch_dict[str(k)]
                                diff_min = np.inf
                                for RO3 in final_epoch:
                 
                                    x3 = RO3.get_x()
                                    y3 = RO3.get_y()
                                    y3_star = m*x3+b
                                    
                                    diff =np.abs(y3_star-y3)**2.
                                    if np.isnan(y3_star):
                                        diff = np.inf
                                    if diff < diff_min:
                                        diff_min = diff
                                diffs.append(diff_min)
                          
                        diffs = sorted(diffs, reverse=True)
                        size = len(diffs)
                        for combo in xrange(size):
                            score = 0
                            for diff in diffs:
                                score += diff     
                            space.append((m,b,score, len(diffs), x1, y1, D, ep1, t))
                            diffs = diffs[1:]
    return space
```

```python
def rejection_sample(space, xbins, ybins,N):
    samples = []
    normed_space = space.copy()
    norm = float(np.max(space))
    normed_space = (normed_space/norm)
    normed_space -= np.min(normed_space)
    total = np.sum(normed_space)
    prob_point = total/N
    I, J = normed_space.shape
    for i in xrange(I):
        for j in xrange(J):
            current_p = normed_space[j,i]
            numpoints = int(current_p/prob_point)
            for n in xrange(numpoints):
                x_sample = np.random.uniform(low=xbins[i], high=xbins[i+1])
                y_sample = np.random.uniform(low=ybins[j], high=ybins[j+1])
                samples.append((x_sample, y_sample))
    return samples
    
```
